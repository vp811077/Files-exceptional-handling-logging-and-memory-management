9. What is the difference between multithreading and multiprocessing?

Ans: The **difference between multithreading and multiprocessing** lies in **how tasks are executed and how system resources are used**.

---

## Multithreading

* Uses **multiple threads** within a **single process**
* Threads **share the same memory space**
* Faster communication between threads
* Affected by **GIL (Global Interpreter Lock)** in Python
* Best for **I/O-bound tasks**

**Example:** File handling, network requests

---

## Multiprocessing

* Uses **multiple processes**
* Each process has its **own memory space**
* True parallel execution
* Not affected by GIL
* Best for **CPU-bound tasks**

**Example:** Data processing, mathematical computations

---

## Key Differences

| Feature           | Multithreading | Multiprocessing |
| ----------------- | -------------- | --------------- |
| Unit of execution | Thread         | Process         |
| Memory            | Shared         | Separate        |
| Speed (CPU tasks) | Limited        | Faster          |
| GIL effect        | Yes            | No              |

---

### **Short exam answer**

> Multithreading runs multiple threads in one process with shared memory, while multiprocessing runs multiple processes with separate memory and true parallelism.

15. How does Python's garbage collection system work?

Ans:**Python's garbage collection system** automatically manages memory by **reclaiming objects that are no longer in use**.

### How it works:

1. **Reference Counting**

   * Every object keeps a **count of references** pointing to it.
   * When the reference count becomes **zero**, the object is **deleted automatically**.

2. **Garbage Collector (GC)**

   * Handles **circular references** that reference counting alone cannot remove.
   * Periodically detects and frees memory occupied by **unreachable objects**.

3. **Modules & Functions**

   * Python provides the `gc` module to **interact with the garbage collector**, e.g., `gc.collect()`.

### **Short exam answer**

> Pythonâ€™s garbage collection system automatically frees memory by deleting objects with zero references and collecting circular references using the garbage collector.

**Example:**

```python
import gc

a = [1, 2, 3]
b = a
del a  # reference count decreases
del b  # object deleted, memory freed
gc.collect()  # optional, forces garbage collection
```

* High-level interface to **create and manage processes**

* Handles **data sharing and synchronization** safely

**Example:**

```python

    print("Process running")

p.join()
```

---

## Key Differences

| ---------------- | -------------------------- | ---------------------------------- |

23. What are the challenges associated with memory management in Python?

Ans:
The **challenges associated with memory management in Python** arise because Python automatically manages memory, but certain situations can still cause issues.

### Key Challenges

1. **Memory leaks**

   * Occur if references to unused objects are not released, especially in **circular references**.

2. **Circular references**

   * Objects referencing each other can prevent reference counts from reaching zero, requiring **garbage collector** intervention.

3. **High memory usage**

   * Large data structures or inefficient code can consume too much memory.

4. **Garbage collection overhead**

   * Automatic garbage collection can sometimes **slow down the program**.

5. **Unpredictable deallocation**

   * Memory is freed when objects are collected, but the exact timing can be **uncertain**, which may affect performance or resource handling.

---

### **Short exam answer**

> Challenges in Python memory management include memory leaks, circular references, high memory usage, garbage collection overhead, and unpredictable deallocation.

25. Why is it important to use multithreading in certain applications?

Ans: **Multithreading is important in certain applications** because it allows a program to **perform multiple tasks concurrently**, improving **performance and responsiveness**.

### Key Reasons

1. **Better performance for I/O-bound tasks**

   * Tasks like file reading/writing, network requests, or database operations can run in parallel while waiting for I/O.

2. **Responsive programs**

   * In GUI or web applications, multithreading prevents the program from freezing during long operations.

3. **Efficient resource utilization**

   * Multiple threads share the same memory, reducing overhead compared to creating multiple processes.

4. **Simplifies program design**

   * Concurrent tasks can be handled within a single process without complex inter-process communication.

---

### **Short exam answer**

> Multithreading is used to run multiple tasks concurrently, improving performance, responsiveness, and efficient resource use in I/O-bound or interactive applications.

**Example:**

```python
import threading
import time

def task(name):
    print(f"Task {name} started")
    time.sleep(2)
    print(f"Task {name} finished")

t1 = threading.Thread(target=task, args=("A",))
t2 = threading.Thread(target=task, args=("B",))

t1.start()
t2.start()
t1.join()
t2.join()
```

This runs **Task A and Task B concurrently**, saving time compared to running them sequentially.

````python
#18.How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

# Create a logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.DEBUG)  # Log all levels DEBUG and above

# Create a rotating file handler (1 MB per file, keep 3 backups)
handler = RotatingFileHandler(
    "app_rotating.log", maxBytes=1_000_000, backupCount=3
)

# Create a formatter and add it to the handler
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Add handler to the logger
logger.addHandler(handler)

# Example log messages
logger.info("This is an info message")
logger.warning("This is a warning message")
logger.error("This is an error message")

````

````python
#19.Write a program that handles both IndexError and KeyError using a try-except block
# Sample list and dictionary
my_list = [10, 20, 30]
my_dict = {"name": "Alice", "age": 25}

try:
    # Accessing a list element that may not exist
    print(my_list[5])

    # Accessing a dictionary key that may not exist
    print(my_dict["address"])

except IndexError:
    print("Error: List index is out of range.")
except KeyError:
    print("Error: Dictionary key does not exist.")

````

````python
#20.How would you open a file and read its contents using a context manager in Python?
file_path = "example.txt"

try:
    with open(file_path, "r") as file:
        content = file.read()  # Read entire file
        print(content)
except FileNotFoundError:
    print(f"Error: The file '{file_path}' does not exist.")

````

````python
#21. Write a Python program that reads a file and prints the number of occurrences of a specific word
file_path = "example.txt"
word_to_count = "Python"

try:
    with open(file_path, "r") as file:
        content = file.read()  # Read entire file
        # Count occurrences of the word (case-sensitive)
        count = content.count(word_to_count)
    print(f"The word '{word_to_count}' occurs {count} times in the file.")
except FileNotFoundError:
    print(f"Error: The file '{file_path}' does not exist.")

````

````python
#22.How can you check if a file is empty before attempting to read its contents?
import os

file_path = "example.txt"

if os.path.exists(file_path) and os.path.getsize(file_path) > 0:
    with open(file_path, "r") as file:
        content = file.read()
        print(content)
else:
    print("The file is empty or does not exist.")

````

````python
#23.Write a Python program that writes to a log file when an error occurs during file handling.
import logging

# Configure logging to write to a file
logging.basicConfig(
    filename="file_errors.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

file_path = "nonexistent_file.txt"

try:
    # Attempt to open and read the file
    with open(file_path, "r") as file:
        content = file.read()
        print(content)
except FileNotFoundError as e:
    logging.error("File not found: %s", e)
    print(f"Error logged: {e}")
except IOError as e:
    logging.error("I/O error occurred: %s", e)
    print(f"Error logged: {e}")

````
